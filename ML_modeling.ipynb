```python
import yfinance as yfin
import numpy as np 
import pandas as pd 
import pandas_datareader.data as pdr
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose

yfin.pdr_override()

# List of stock symbols
# S&P 500 index
ticker = '^GSPC'

# Set the date range for the historical data
start_date = '2015-01-01'
end_date = '2021-01-01'

# Download historical stock data for the S&P 500 with interval of 1 month

stocks = pdr.get_data_yahoo(ticker, start_date, end_date,interval='1mo').Close

```

```python
stocks.plot()
```

```python
# Plot  the components of a time series by
# seasonal_decompose function from statsmodels

seasonal_decompose(stocks, period=12).plot()

plt.show()
```

The second panel shows the trend and third shows seasonality.

The trend can be understood using the Correlation function and auto-correlation functions.
